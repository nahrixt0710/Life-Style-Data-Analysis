```python
import pandas as pd

df = pd.read_csv("../data/data.csv")
```

### 1 . Chọn feature sử dụng

Sau khi EDA và thực hiện tính toán pearson corelation thì ta đã chọn ra được 6 cột

```python
chosen_col = ['BMI', 'Weight (kg)', 'Calories', 'Height (m)', 'Burns Calories (per 30 min)', 'Water_Intake (liters)']

X = df[chosen_col]
y = df['Fat_Percentage']
X
```

### 2 . Thực hiện scaling (Standard Scaler)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled
```

```python
df_new = pd.concat([X_scaled, y], axis=1)
df_new
```

```python
df_new.to_csv('../data/after_preprocess_no_categorical.csv', index=False)
```
