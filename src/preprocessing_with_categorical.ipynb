### 1 . Chọn feature sử dụng

```python
import numpy as np

chosen_col = ['BMI', 'Weight (kg)', 'Calories', 'Height (m)', 'Burns Calories (per 30 min)', 'Water_Intake (liters)']

X_num = df[chosen_col]
y = df['Fat_Percentage']

X_cate = df.select_dtypes(exclude=[np.number])
```

Các biến nên drop : 

    - meal_name : giá trị chỉ toàn là Other
    - Name of Exercise : đây chỉ là tên bài tập, không có giá trị cho việc dự đoán
    - Benefit : Đây là hiệu quả của bài tập, phần nhiều đã được liệt kê trong các biến như Target Muscle Group hay Equipment Needed
    - Workout : Tương tự như biến Name of Exercise

```python
X_cate = X_cate.drop("meal_name", axis=1)
X_cate = X_cate.drop("Name of Exercise", axis=1)
X_cate = X_cate.drop("Benefit", axis=1)
X_cate = X_cate.drop("Workout", axis=1)
```

Các cột còn lại được biến đổi bằng 3 loại biến đổi :

    - LabelEncoder : Gender (vì chỉ có 2 giá trị)
    - OridnalEncoder : Difficulty Level, Burns_Calories_Bin (các giá trị có thứ bậc)
    - OnehotEncoder : Các biến còn lại (là các dạng biến phân loại thông thường)

Với Target Muscle Group và Equipment Needed sẽ xử lí khác

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer

def onehot_encode(col):
    df_ = X_cate[[col]]
    encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
    ct = ColumnTransformer(
        transformers=[("ohe", encoder, [col])],
        remainder="passthrough" 
    )
    x_one_hot = ct.fit_transform(df_)
    feature_names = ct.get_feature_names_out()
    df_one_hot = pd.DataFrame(x_one_hot, columns=feature_names)

    return df_one_hot
    

def oridinal_encode(col, ord):
    df_ = X_cate[[col]]
    encoder = OrdinalEncoder(categories=[ord])
    return encoder.fit_transform(df_)
```

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder
le = LabelEncoder()
X_cate['Gender'] = le.fit_transform(X_cate['Gender'])

# OrdinalEncoder
X_cate["Difficulty Level"] = oridinal_encode("Difficulty Level", ['Beginner', 'Intermediate', 'Advanced'])
X_cate["Burns_Calories_Bin"] = oridinal_encode("Burns_Calories_Bin",  ['Low', 'Medium', 'High', 'Very High'])

# OnehotEncoder
ohe_list = ["Workout_Type", "meal_type", "diet_type", "cooking_method", "Body Part", "Type of Muscle"]
for ohe_col in ohe_list:
    ohe = onehot_encode(ohe_col)
    X_cate = X_cate.drop(ohe_col, axis=1)
    X_cate = pd.concat([X_cate, ohe], axis=1)
```

#### Target Muscle Group và Equipment Needed 

Với Target Muscle Group thì đầu tiên cần phải tách các nhóm ra và tạo thành các cột thuộc tính muscle 

```python
muscles = []
for muscle in X_cate['Target Muscle Group']:
    muscle = muscle.split(',')
    for m in muscle:
        m = m.strip().lower()
        if m not in muscles:
            muscles.append(f"{m}")

print(muscles)

# Tạo cột toàn 0 
X_cate[muscles] = 0

for i, muscle in enumerate(X_cate['Target Muscle Group']):
    muscle = muscle.lower()
    for m in muscles:
        if m in muscle:
            X_cate.at[i, m]  = 1

X_cate
```

Equipment Needed

```python
equipments = []
for equip in X_cate['Equipment Needed']:
    equip = equip.replace("or ", ",")
    equip = equip.split(',')
    for e in equip:
        e = e.strip().lower()
        if e not in equipments:
            equipments.append(f"{e}")

print(equipments)

# Tạo cột toàn 0 
X_cate[equipments] = 0

for i, equip in enumerate(X_cate['Equipment Needed']):
    equip = equip.lower()
    for e in equipments:
        if e in equip:
            X_cate.at[i, e]  = 1

X_cate
```
